<p>Дано множество наблюдений (файл "Reg_A5.csv"), регрессор - линейная регрессия. Найти степень полинома с минимальной ошибкой на проверочном подможестве, определить среднеквадратическую ошибку на тестовом подмножестве? (Степень полинома от 1 до 25)</p>
<p>
Построить графики:<br>
1) зависимости <b>среднеквадратической ошибки</b> от степени полинома для обучающего и проверочного подмножеств<br>
2) зависимости <b>R-квадрата</b> от степени полинома для обучающего и проверочного подмножеств<br>
3) функцию регрессии (наилучший случай) + исходные данные</p>

```python
from sklearn import datasets
from scipy import stats
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_predict, train_test_split
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.model_selection import cross_val_predict, train_test_split

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from numpy.polynomial.polynomial import polyval

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.linear_model import RidgeCV

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
clrMap = ListedColormap(["blue", "red", "green"])

%matplotlib inline
```

<p>1. Загрузка исходных данных</p>

```python
df = pd.read_csv("data/Reg_A5.csv")

x = df.X.values
y = df.Y.values

plt.title("Initial Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, "o")
plt.grid(True)
plt.show()
```

<p>2. Формирование обучающего (train), проверочного (val) и тестового (test) подмножеств двумя способами</p>

```python
# Способ 1
x_train_init, x_test, y_train_init, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
x_train, x_val, y_train, y_val = train_test_split(x_train_init, y_train_init, test_size=0.25, random_state=0)

# Способ 2

    # Комментарий преподавателя:
    # Воспользоваться KFold с параметрами n_splits=4, shuffle=True, random_state=0
    # Можно использовать функцию cross_validate()

# Использована функция cross_validate(), которая разбивает внутри себя
```

<p>3. Обучение</p>

```python
    # Комментарий преподавателя:
    # LinearRegression с нормализацией X

    # Для способов 1 и 2

    # Для способа 2
    # Для каждой степени полинома ошибка обучения:
    # MeanErrorTrain = 1/4*(ErrorTrainFold_1+ErrorTrainFold_2+ErrorTrainFold_3+ErrorTrainFold_4)
    # или Можно использовать функцию cross_validate()

    # Для каждой степени полинома ошибка на проверочном подмножестве:
    # MeanErrorValidation = 1/4*(ErrorValFold_1+ErrorValFold_2+ErrorValFold_3+ErrorValFold_4)
    # или Можно использовать функцию cross_validate()
```

```python
x_train = x_train.reshape(-1, 1)
x_val = x_val.reshape(-1, 1)

x_train_init = x_train_init.reshape(-1,1)

pipeline = Pipeline([("plF", PolynomialFeatures(degree=1)), ("lr", LinearRegression(fit_intercept=True))])

pipeline = pipeline.fit(x_train, y_train)
pipeline
f_pred = lambda x : polyval(x, pipeline.named_steps["lr"].coef_) + pipeline.named_steps["lr"].intercept_
```

```python
# Проверка коэффициентов
print("Коэффициенты: ", pipeline.named_steps["lr"].coef_)
print("Пересечение: ", pipeline.named_steps["lr"].intercept_)
```

<p>4. Графики зависимости среднеквадратической ошибки от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2
from sklearn.metrics import mean_squared_error, r2_score

mse_train = []
mse_val = []
cross_val_mse_train = []
cross_val_mse_val = []

for i in range (1,26):
    # Оформляем общий вид модели
    pipeline = Pipeline([("plF", PolynomialFeatures(degree=i)), ("lr", LinearRegression(fit_intercept=True))])
    
    #Обучаем
    pipeline = pipeline.fit(x_train, y_train)
    f_pred = lambda x : polyval(x, pipeline.named_steps["lr"].coef_) + pipeline.named_steps["lr"].intercept_
    mse_train.append(mean_squared_error(y_train, f_pred(x_train).flatten()))
    mse_val.append(mean_squared_error(y_val, f_pred(x_val).flatten()))
    
    scores = cross_validate(pipeline, x_train_init, y_train_init, cv=4, scoring=('r2', 'neg_mean_squared_error'), 
                            return_train_score=True)
    cross_val_mse_train.append((-1) * np.average(scores['train_neg_mean_squared_error']))
    cross_val_mse_val.append((-1) * np.average(scores['test_neg_mean_squared_error']))
    

plt.figure(figsize=[5, 3])

plt.title("Error")
plt.xlabel("Poly Degree")
plt.ylabel("Error")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26), mse_train,"b-o")
plt.plot(np.arange(1, 26),mse_val, "r-o")
plt.grid(True)

plt.figure(figsize=[5, 3])
plt.title("Cross Validation Error")
plt.xlabel("Poly Degree")
plt.ylabel("Error")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26),cross_val_mse_train, "b-o")
plt.plot(np.arange(1, 26),cross_val_mse_val, "r-o")
plt.grid(True)
```

<p>5. Графики зависимости R-квадрата от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2

r2_train = []
r2_val = []
cross_val_r2_train = []
cross_val_r2_val = []

for i in range (1,26):
    pipeline = Pipeline([("plF", PolynomialFeatures(degree=i)), ("lr", LinearRegression(fit_intercept=True))])
    
    pipeline = pipeline.fit(x_train, y_train)
    f_pred = lambda x : polyval(x, pipeline.named_steps["lr"].coef_) + pipeline.named_steps["lr"].intercept_
    r2_train.append(pipeline.score(x_train, y_train))
    r2_val.append(pipeline.score(x_val, y_val))
    
    scores = cross_validate(pipeline, x_train_init, y_train_init, cv=4, scoring=('r2', 'neg_mean_squared_error'), 
                            return_train_score=True)
    cross_val_r2_train.append(np.average(scores['train_r2']))
    cross_val_r2_val.append(np.average(scores['test_r2']))
    
plt.figure(figsize=[5, 3])
plt.title("R-squared")
plt.xlabel("Poly Degree")
plt.ylabel("R-squared")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26),r2_train, "b-o")
plt.plot(np.arange(1, 26),r2_val, "r-o")
plt.grid(True)

plt.figure(figsize=[5, 3])
plt.title("Сross Validation R-squared")
plt.xlabel("Poly Degree")
plt.ylabel("R-squared")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26),cross_val_r2_train, "b-o")
plt.plot(np.arange(1, 26),cross_val_r2_val, "r-o")
plt.grid(True)

# Ваш код
```

<p>6. Выбор степени полинома + повторное обучение модели на train_init</p>

```python
# Для способов 1 и 2
# np.argmin(mse_train)+1)- идея в том, что у нас элементы массива расположены в соответствии в порядком полинома
# но индексация в массиве начинается с нуля, поэтому прибавляем 1
# поэтому поиск наименьшего элемента в итоге возвращает степень полинома

print ("Степень полинома с минимальной ошибкой: ", np.argmin(mse_val)+1)
print ("Степень полинома с максимальным коэффицентом детерминации: ", np.argmax(r2_val)+1)

print ("Степень полинома с минимальной ошибкой при кросс-валидации: ", np.argmin(cross_val_mse_val)+1)
print ("Степень полинома с максимальным коэффицентом детерминации при кросс-валидации: ", np.argmax(cross_val_r2_val)+1)
```

```python
# Обучение на train_init
x_train_init = x_train_init.reshape(-1, 1)

pipeline = Pipeline([("plF", PolynomialFeatures(degree=3)), ("lr", LinearRegression(fit_intercept=True))])
pipeline = pipeline.fit(x_train_init, y_train_init)
f_pred = lambda x : polyval(x, pipeline.named_steps["lr"].coef_) + pipeline.named_steps["lr"].intercept_
```

<p>7. Cреднеквадратическая ошибка на тестовом подмножестве</p>

```python
# Для способов 1 и 2
x_test = x_test.reshape(-1, 1)

MSE = mean_squared_error(y_test, f_pred(x_test).flatten())

print("Среднеквадратичная ошибка: ", MSE)
```

<p>3. Обучение</p>

```python
    # Комментарий преподавателя:
    # Параметры классификатора
    # penalty="l2", fit_intercept=True, max_iter=x, C=1e5, solver="liblinear", random_state=12345

    # Для способов 1 и 2

pf = PolynomialFeatures(degree=2)

logr_model = LogisticRegression(penalty="l2", fit_intercept=True, max_iter=100, C=1e5, multi_class="ovr",
                                solver="liblinear", random_state=12345)

x_train_poly = pf.fit_transform(x_train)
logr_model.fit(x_train_poly, y_train)
```

<p>4. Графики зависимости доли правильных классификаций от степени полинома для обучающего и проверочного подмножеств</p>

```python
# Для способов 1 и 2
import warnings
warnings.filterwarnings('ignore')

score_train_array = []
score_val_array = []
cross_val_train_array = []
cross_val_val_array = []

f_pred = lambda x : logr_model.predict(x)

for i in range (1,26):
    pf = PolynomialFeatures(degree=i)
    
    x_train_poly = pf.fit_transform(x_train)
    logr_model.fit(x_train_poly, y_train)
    score_train_array.append(logr_model.score(x_train_poly, y_train))
    
    x_val_poly = pf.fit_transform(x_val)
    logr_model.fit(x_val_poly, y_val)
    score_val_array.append(logr_model.score(x_val_poly, y_val))
    
    x_train_init_poly = pf.fit_transform(x_train_init)
    scores = cross_validate(logr_model, x_train_init_poly, y_train_init, cv=4, return_train_score=True)
    cross_val_train_array.append(np.average(scores['train_score']))
    cross_val_val_array.append(np.average(scores['test_score']))


plt.figure(figsize=[5, 3])
plt.title("Score")
plt.xlabel("Poly Degree")
plt.ylabel("Score")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26),score_train_array, "-o", color = 'b')
plt.plot(np.arange(1, 26),score_val_array, "-o", color = 'r')
plt.grid(True)

plt.figure(figsize=[5, 3])
plt.title("Cross Validation Score")
plt.xlabel("Poly Degree")
plt.ylabel("Score")
plt.xticks(np.arange(1, 25, step=5)) 
plt.plot(np.arange(1, 26),cross_val_train_array, "-o", color = 'b')
plt.plot(np.arange(1, 26),cross_val_val_array, "-o", color = 'r')
plt.grid(True)
```

```python
# np.argmax(score_train_array)+1) - идея в том, что у нас элементы массива расположены в соответствии в порядком полинома
# но индексация в массиве начинается с нуля, поэтому прибавляем 1
# поэтому поиск наименьшего элемента в итоге возвращает степень полинома

print ("Максимальная доля правильных классификаций на обучающей выборке соответствует степени полинома: ", 
       np.argmax(score_train_array)+1)
print ("Максимальная доля правильных классификаций на проверочной выборке соответствует степени полинома: ", 
       np.argmax(score_val_array)+1)
print ("Максимальная доля правильных классификаций на обучающей выборке при кросс-валидации соответствует степени полинома: ", 
       np.argmax(cross_val_train_array)+1)
print ("Максимальная доля правильных классификаций на проверочной выборке при кросс-валидации соответствует степени полинома: ", 
       np.argmax(cross_val_val_array)+1)
```

<p>5. Графики зависимости доли правильных классификаций от количества итераций для обучающего (train) и проверочного подмножеств для наилучшего случая - изменение параметра max_iter</p>

<p>2. Обучение с кросс-валидацией (см. задачу 1 способ 2) и с использованием Ridge при различных значениях коэф. регуляризации</p>

```python
alphas = np.append([0.0], np.logspace(-8, 1, 20, base=10))
alphas
```

```python
    # Комментарий преподавателя:
    # Регрессор - Ridge
    # Параметры: a fit_intercept=True, normalize=True, random_state=12345
    # Здесь и далее СТЕПЕНЬ ПОЛИНОМА - 16!!!

x = x.reshape(-1, 1)
x_init, x_val, y_init, y_val = train_test_split(x, y, test_size=0.3, random_state=12345)
x_train, x_test, y_train, y_test = train_test_split(x_init, y_init, test_size=0.25, random_state=12345)

# Коэф. регуляризации
alphas = np.append([0.0], np.logspace(-8, 1, 20, base=10))

pl_ridge = Pipeline([("plF", PolynomialFeatures(degree=16)), ("lr", RidgeCV(alphas=alphas,
                                                                            normalize=True,
                                                                            fit_intercept=True, cv=4, 
                                                                            store_cv_values=False))])
pl_ridge.fit(x_train, y_train)

print("Коэффициенты: ", pl_ridge.named_steps["lr"].coef_)
print("Пересечение: ", pl_ridge.named_steps["lr"].intercept_)
print("Альфа: ", pl_ridge.named_steps["lr"].alpha_ )

f_x = lambda x : polyval(x, pl_ridge.named_steps["lr"].coef_) + pl_ridge.named_steps["lr"].intercept_
```

<p>3. Графики зависимости среднеквадратической ошибки от коэффициента регуляризации для обучающего и проверочного подмножеств</p>

```python
mse_train = []
mse_val = []

for alpha in alphas:
    pl_ridge = Pipeline([("plF", PolynomialFeatures(degree=16)), ("lr", RidgeCV(alphas=[alpha],
                                                                            normalize=True,
                                                                            fit_intercept=True, cv=4, 
                                                                            store_cv_values=False))])
    pl_ridge.fit(x_train, y_train)
    f_pred = lambda x : polyval(x, pl_ridge.named_steps["lr"].coef_) + pl_ridge.named_steps["lr"].intercept_
    mse_train.append(mean_squared_error(y_train, f_pred(x_train).flatten()))
    mse_val.append(mean_squared_error(y_val, f_pred(x_val).flatten()))
    
plt.title("Error")
plt.xlabel("Regularization coef")
plt.ylabel("Error")
plt.xscale('log')
plt.plot(alphas, mse_train, "b-o")
plt.plot(alphas, mse_val,"r-o")
plt.grid(True)
```

<p>4. Графики зависимости R-квадрата от коэффициента регуляризации для обучающего и проверочного подмножеств</p>

```python
r2_train = []
r2_val = []

for alpha in alphas:
    pl_ridge = Pipeline([("plF", PolynomialFeatures(degree=16)), ("lr", RidgeCV(alphas=[alpha],
                                                                            normalize=True,
                                                                            fit_intercept=True, cv=4, 
                                                                            store_cv_values=False))])
    pl_ridge.fit(x_train, y_train)
    f_pred = lambda x : polyval(x, pl_ridge.named_steps["lr"].coef_) + pl_ridge.named_steps["lr"].intercept_
    r2_train.append(pl_ridge.score(x_train, y_train))
    r2_val.append(pl_ridge.score(x_val, y_val))
    
plt.title("R-squared")
plt.xlabel("Poly Degree")
plt.ylabel("R-squared")
plt.xscale('log')
plt.plot(alphas, r2_train, "b-o")
plt.plot(alphas, r2_val, "r-o")
plt.grid(True)
plt.show()
```

<p>5. Найти коэф. рег. с минимальной ошибкой на val</p>

```python
bestAlpha = alphas[np.argmin(mse_val)]
print("Лучший коэффициент регуляризации:", bestAlpha)
```

<p>6. Обучение на init с использованием Ridge и параметром alpha=(значение из пункта 5)</p>

```python
pl_ridge = Pipeline([("plF", PolynomialFeatures(degree=16)), ("lr", RidgeCV(alphas=[bestAlpha],
                                                                            normalize=True,
                                                                            fit_intercept=True, cv=4, 
                                                                            store_cv_values=False))])
pl_ridge.fit(x_init, y_init)
```

<p>7. Определить ошибку на тестовом подмножестве</p>

```python
f_pred = lambda x : polyval(x, pl_ridge.named_steps["lr"].coef_) + pl_ridge.named_steps["lr"].intercept_
MSE = mean_squared_error(y_test, f_pred(x_test))
print("Среднеквадратичная ошибка:", MSE)
```
